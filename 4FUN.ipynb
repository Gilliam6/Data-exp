```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with plt.xkcd():
    fig = plt.figure()
    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylim([-30, 10])

    data = np.ones(200)
    data[170:] -= np.arange(30)

    ax.annotate(
        'THE DAY I REALIZED\nI COULD LISTEN\n "Im A CREEP"\nWHENEVER I WANTED',
        xy=(170, 1), arrowprops=dict(arrowstyle='->'), xytext=(20, -20))

    ax.plot(data)

    ax.set_xlabel('time')
    ax.set_ylabel('my mental health')
    fig.text(
        0.5, 0.05,
        '',
        ha='center')
    plt.savefig('pacani_smotrite_chto_prines.png')
```
